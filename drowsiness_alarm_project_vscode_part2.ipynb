```python
from pygame import mixer

mixer.init()

sound=mixer.Sound(r"C:\Users\viswa\OneDrive\Desktop\alarm.wav")






# lets use cv2 and capture video and in that video we give a rectangle to face and small rectangle to eyes

import cv2
import numpy as np

# face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# eye detection
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')







# capture video
cap = cv2.VideoCapture(0)
score=0

while True:

    # read the frame
    ret, frame = cap.read()
    # print(frame.shape)
    height,width=frame.shape[:2]  #frame is a numpy array has height ,width and channel as list thats why we are taking height and width[0,1]

    # create rectangle to show text
    cv2.rectangle(frame,(0,height-50),(200,height),(0,0,0),thickness=cv2.FILLED)

    # convert into gray scale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # detect faces
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # draw rectangle around the face
    for (x, y, w, h) in faces:

        # draw rectangle around the face
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

        # detect eyes
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]

        # detect eyes
        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 3)

        # draw rectangle around the eyes
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh),(0, 255, 0), 2)
            # pre processing the image
            eye=frame[ey:ey+eh,ex:ex+ew]
            eye=cv2.resize(eye,(150,150))
            # eye=cv2.cvtColor(eye,cv2.COLOR_BGR2GRAY)
            eye=eye/255
            eye=eye.reshape(150,150,3)

            eye=np.expand_dims(eye,axis=0)

            # predict the image
            prediction=model.predict(eye)
            
            if prediction>0.4:
                cv2.putText(frame,'open',(10,height-20),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2,cv2.LINE_AA)
                score=score-1
                                




            elif prediction<0.4:
                cv2.putText(frame,'closed',(10,height-20),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2,cv2.LINE_AA)
                score=score+1
                if score>15:
                    try:
                        sound.play()
                    except:
                        pass
                    # cv2.putText(frame,'Drowsy',(100,height-20),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2,cv2.LINE_AA)

                


            if score<0:
                score=0

            cv2.putText(frame,'score:'+str(score),(100,height-20),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2,cv2.LINE_AA)




 











            

    # display the frame
    cv2.imshow('frame', frame)

    # press q to exit
    if cv2.waitKey(1) == ord('q'):
        break

# release the capture
cap.release()

# destroy all windows
cv2.destroyAllWindows()





    
```
